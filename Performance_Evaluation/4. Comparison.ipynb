# **Import lib**

```python
import pandas as pd
from matplotlib import pyplot as plt
```

# **Load performance data**

You may need to upload 'performance.csv' to '/content/performance.csv' first if you are using google colab

```python
performance = pd.read_csv('/content/performance.csv', header=0, index_col='futureStep').drop(['Unnamed: 0'], axis='columns')
performance
```

# **Plot Training time, Prediction time and RMSE**

```python
plt.figure(figsize=(15, 5))
plt.plot(performance['LSTM_train_time'], label='LSTM training time')
plt.plot(performance['VAR_train_time'], label='VAR training time')
plt.plot(performance['Prophet_train_time'], label='Prophet training time')
plt.title("Training time")
plt.xlabel('future step')
plt.ylabel('second')
plt.legend()
plt.show()

plt.figure(figsize=(15, 5))
plt.plot(performance['LSTM_pred_time'], label='LSTM prediction time')
plt.plot(performance['VAR_pred_time'], label='VAR prediction time')
plt.plot(performance['Prophet_pred_time'], label='Prophet prediction time')
plt.title("Prediction time")
plt.xlabel('future step')
plt.ylabel('second')
plt.legend()
plt.show()

plt.figure(figsize=(15, 5))
plt.plot(performance['LSTM_RMSE'], label='LSTM RMSE')
plt.plot(performance['VAR_RMSE'], label='VAR RMSE')
plt.plot(performance['Prophet_RMSE'], label='Prophet RMSE')
plt.title("RMSE")
plt.xlabel('future step')
plt.ylabel('RMSE')
plt.legend()
plt.show()
```

Note: The FB Prophet seems having the best accuracies, but to forecast data by FB Prophet for more than one future steps, we need to predict all variables first. One output y is only corresponding to one row of variable.
